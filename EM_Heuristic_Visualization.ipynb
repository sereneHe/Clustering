```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

Y = pd.read_csv('HeartRate_MIT_Pred.csv',sep=',',header='infer')

y_1 = list(Y['Health_1'])
y_2 = list(Y['Health_2'])
y_3 = list(Y['Health_3'])
y_4 = list(Y['Disease_1'])
y_5 = list(Y['Disease_2'])
y_6 = list(Y['Disease_3'])
y_7 = list(Y['Prediction_Health'])
y_8 = list(Y['Prediction_Disease'])
#y_7 = rolling_mean_H
#y_8 = rolling_mean_D
x = range(len(Y))


plt.figure(figsize=(20,8),dpi=80)

plt.plot(x,y_1,label="Normal Sinus Rhythm_Patient1",
         color="lightskyblue",
         linestyle=':',
         linewidth=5,
         alpha=0.8)

plt.plot(x,y_2,label="Normal Sinus Rhythm_Patient2",
         color="cyan",
         linestyle=':',
         linewidth=5,
         alpha=0.8)
plt.plot(x,y_3,label="Normal Sinus Rhythm_Patient3",
         color="sandybrown",
         linestyle=':',
         linewidth=5,
         alpha=0.8)
plt.plot(x,y_7,label="EM Heuristic_Prediction",
         color="red",
         linestyle='-',
         linewidth=5,
         alpha=0.8)
_xtick_labels = [i for i in x]
plt.xticks(x,_xtick_labels )

plt.grid(alpha=0.4)
x_major_locator = plt.MultipleLocator(20)
ax = plt.gca()
plt.xlim(0,620)
ax.xaxis.set_major_locator(x_major_locator)
plt.xlabel('Time(sec*2)', fontsize=14)
plt.ylabel('Heart rate(bpm)', fontsize=14)
plt.xlim(0,640)
plt.legend(loc="lower right", fontsize=9)

plt.show()
           
           
           
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_4,label="Ventricular Fibrillation_Patient1",
         color="paleturquoise",
         linestyle=':',
         linewidth=5,
         alpha=0.8)
plt.plot(x,y_5,label="Ventricular Fibrillation_Patient2",
         color="lightgreen",
         linestyle=':',
         linewidth=5,
         alpha=0.8)
plt.plot(x,y_6,label="Ventricular Fibrillation_Patient3",
         color="violet",
         linestyle=':',
         linewidth=5,
         alpha=0.8)
plt.plot(x,y_8,label="EM Heuristic_Prediction",
         color="navy",
         linestyle='-',
         linewidth=5,
         alpha=0.8)

#_xtick_labels = ["Day{}".format(i) for i in x]
_xtick_labels = [i for i in x]
plt.xticks(x,_xtick_labels )

plt.grid(alpha=0.4)
x_major_locator = plt.MultipleLocator(20)
ax = plt.gca()
plt.xlim(0,640)
ax.xaxis.set_major_locator(x_major_locator)
plt.xlabel('Time(sec*2)', fontsize=14)
plt.ylabel('Heart rate(bpm)', fontsize=14)

plt.legend(loc="lower right", fontsize=9.5)

plt.show()
```
